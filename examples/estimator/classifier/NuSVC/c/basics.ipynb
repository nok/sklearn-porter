# sklearn-porter

Repository: https://github.com/nok/sklearn-porter

## NuSVC

Documentation: [sklearn.svm.NuSVC](http://scikit-learn.org/stable/modules/generated/sklearn.svm.NuSVC.html)

### Loading data:

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
X = iris_data.data
y = iris_data.target

print(X.shape, y.shape)
```

### Train classifier:

```python
from sklearn import svm

clf = svm.NuSVC(gamma=0.001, kernel='rbf', random_state=0)
clf.fit(X, y)
```


### Transpile classifier:

```python
%%time

from sklearn_porter import Porter

porter = Porter(clf, language='c')
output = porter.export()

print(output)
```

### Run classification in C:

Save the transpiled estimator:

```python
with open('nusvc.c', 'w') as f:
    f.write(output)
```

Compiling:

```python
%%bash

gcc nusvc.c -std=c99 -lm -o nusvc
```

Prediction:

```python
%%bash

./nusvc 1 2 3 4
```
