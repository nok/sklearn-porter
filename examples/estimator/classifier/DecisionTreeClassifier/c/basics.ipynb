# sklearn-porter

Repository: https://github.com/nok/sklearn-porter

## DecisionTreeClassifier

Documentation: [sklearn.tree.DecisionTreeClassifier](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

### Loading data:

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
X = iris_data.data
y = iris_data.target

print(X.shape, y.shape)
```

### Train classifier:

```python
from sklearn.tree import tree

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```


### Transpile classifier:

```python
%%time

from sklearn_porter import Porter

porter = Porter(clf, language='c')
output = porter.export()

print(output)
```

### Run classification in C:

Save the transpiled estimator:

```python
with open('tree.c', 'w') as f:
    f.write(output)
```

Compiling:

```python
%%bash

gcc tree.c -std=c99 -lm -o tree
```

Prediction:

```python
%%bash

./tree 1 2 3 4
```
